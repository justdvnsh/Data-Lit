```
df.dropna()
df_2 = df.groupby(['Educational Level of the Women']).mean()
df_2
```

```
from bokeh.io import show, output_file, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_notebook()

colors_tfr = ['#ffa7b6', '#ff8ea1', '#ff748c', '#ff5b77', '#ff4262', '#ff284d', '#ff0f38', '#f5002a']
colors_15_19 = ['#4bf3ab', '#33f1a0', '#1bf095', '#0fe288', '#0dca7a', '#0cb26b', '#0a9a5d', '#08824f']
colors_20_24 = ['#40cedb', '#2bc9d7', '#25b7c4', '#21a3ae', '#1d8f99', '#197b84', '#15676e', '#115359']
colors_25_29 = ['#ffff00', '#e6e600', '#cccc00', '#b3b300', '#9a9a00', '#808000', '#676700', '#4e4e00']
colors_30_34 = ['#ba61d8', '#b14dd3', '#a839ce', '#9a2fbe', '#892aaa', '#792595', '#682081', '#581b6d']
colors_35_39 = ['#d1dd5f', '#cbd94a', '#c5d435', '#b7c62a', '#a4b126', '#909c21', '#7d871d', '#6a7218']

ed_level = np.array(df_2.index)
tfr = df_2['Total Fertility Rate']

tfr = sorted(tfr)

tfr_15_19 = sorted(df_2['Fertility Rate of Age Group 15-19'])
tfr_20_24 = sorted(df_2['Fertility Rate of Age Group 20-24'])
tfr_25_29 = sorted(df_2['Fertility Rate of Age Group 25-29'])
tfr_30_34 = sorted(df_2['Fertility Rate of Age Group 30-34'])
tfr_35_39 = sorted(df_2['Fertility Rate of Age Group 35-39'])
tfr_40_44 = sorted(df_2['Fertility Rate of Age Group 40-44'])
tfr_45_49 = sorted(df_2['Fertility Rate of Age Group 45-49'])


print(tfr)


def make_plot(col, sorted_tfr, colors):
  
  sorted_ed_level = []
  for tfr in sorted_tfr:
    sorted_ed_level.append(df_2.index[df_2[col] == tfr][0])

  print(sorted_ed_level)

  source = ColumnDataSource(data=dict(sorted_ed_level=sorted_ed_level, sorted_tfr=sorted_tfr, color=colors))

  p = figure(x_range=sorted_ed_level, plot_height=500, plot_width=900, title="Fertility rate based on Education Level" + col,
             toolbar_location=None, tools="")

  p.vbar(x='sorted_ed_level', top='sorted_tfr', width=0.9, color='color', legend='sorted_ed_level', source=source)

  p.xgrid.grid_line_color = None
  p.y_range.start = 0
  p.legend.orientation = "vertical"
  p.legend.location = "top_left"

  show(p)

make_plot('Total Fertility Rate', tfr, colors_tfr)
make_plot('Fertility Rate of Age Group 15-19', tfr_15_19, colors_15_19)
make_plot('Fertility Rate of Age Group 20-24', tfr_20_24, colors_20_24)
make_plot('Fertility Rate of Age Group 25-29', tfr_25_29, colors_25_29)
make_plot('Fertility Rate of Age Group 30-34', tfr_30_34, colors_30_34)
make_plot('Fertility Rate of Age Group 35-39', tfr_35_39, colors_35_39)
make_plot('Fertility Rate of Age Group 40-44', tfr_40_44, colors_15_19)
make_plot('Fertility Rate of Age Group 45-49', tfr_45_49, colors_20_24)
```

So we can see that , illiterate women tend to have higher fertility rate than literate women , in all the age groups of their fertility lifetime . Which makes complete sense becuase - This is what who suggests in its report

> Investments in womenâ€™s education have proven especially effective at lowering fertility rates, since better educated women tend to marry later and have fewer, healthier children. Studies on every continent show that as literacy rates rise, especially those of women, income levels, nutrition levels, and child survival rates rise as well. A study using data from 219 countries from 1970 to 2009 found that, for every one additional year of education for women of reproductive age, child mortality decreased by 9.5 per cent.

```
df_3 = df.groupby(['Year']).mean()
df_3
```

```
output_notebook()

year = np.array(df_3.index)
tfr = df_3['Total Fertility Rate']
tfr_15_19 = df_3['Fertility Rate of Age Group 15-19']
tfr_20_24 = df_3['Fertility Rate of Age Group 20-24']
tfr_25_29 = df_3['Fertility Rate of Age Group 25-29']
tfr_30_34 = df_3['Fertility Rate of Age Group 30-34']
tfr_35_39 = df_3['Fertility Rate of Age Group 35-39']
tfr_40_44 = df_3['Fertility Rate of Age Group 40-44']
tfr_45_49 = df_3['Fertility Rate of Age Group 45-49']

def make_plot_year(col, sorted_tfr):

  source = ColumnDataSource(data=dict(year=year, tfr=sorted_tfr))

  p = figure(plot_height=500, plot_width=900, title="Fertility rate based on Year " + col)

  p.line('year', 'tfr', source=source, line_width=3)

  p.xgrid.grid_line_color = None
  #p.y_range.start = 0

  show(p)
  
make_plot_year('Total Fertility Rate', tfr)
make_plot_year('Fertility Rate of Age Group 15-19', tfr_15_19)
make_plot_year('Fertility Rate of Age Group 20-24', tfr_20_24)
make_plot_year('Fertility Rate of Age Group 25-29', tfr_25_29)
make_plot_year('Fertility Rate of Age Group 30-34', tfr_30_34)
make_plot_year('Fertility Rate of Age Group 35-39', tfr_35_39)
make_plot_year('Fertility Rate of Age Group 40-44', tfr_40_44)
make_plot_year('Fertility Rate of Age Group 45-49', tfr_45_49)
```

Here too, we can see that in the span of 5 years i.e. from 2007 to 2011 , the fertility rates have gone down , since the females have become more literate over those years and tend to marry later and use birth control methods for the same .

```
df_4 = df.groupby(['Area']).mean()
df_4.drop(['Total'], axis=0, inplace=True)
df_4
```

```
output_notebook()

area = np.array(df_4.index)
tfr = df_4['Total Fertility Rate']

tfr = sorted(tfr)

tfr_15_19 = sorted(df_4['Fertility Rate of Age Group 15-19'])
tfr_20_24 = sorted(df_4['Fertility Rate of Age Group 20-24'])
tfr_25_29 = sorted(df_4['Fertility Rate of Age Group 25-29'])
tfr_30_34 = sorted(df_4['Fertility Rate of Age Group 30-34'])
tfr_35_39 = sorted(df_4['Fertility Rate of Age Group 35-39'])
tfr_40_44 = sorted(df_4['Fertility Rate of Age Group 40-44'])
tfr_45_49 = sorted(df_4['Fertility Rate of Age Group 45-49'])


def make_plot_area(col, sorted_tfr, colors):
  
  sorted_area = []
  for tfr in sorted_tfr:
    sorted_area.append(df_4.index[df_4[col] == tfr][0])

  print(sorted_area)

  source = ColumnDataSource(data=dict(sorted_area=sorted_area, sorted_tfr=sorted_tfr, color=colors))

  p = figure(x_range=sorted_area, plot_height=500, plot_width=500, title="Fertility rate based on Area " + col,
             toolbar_location=None, tools="")

  p.vbar(x='sorted_area', top='sorted_tfr', width=0.7, color='color', legend='sorted_area', source=source)

  p.xgrid.grid_line_color = None
  p.y_range.start = 0
  p.legend.orientation = "vertical"
  p.legend.location = "top_left"

  show(p)

make_plot_area('Total Fertility Rate', tfr, [colors_tfr[0],colors_tfr[-1]])
make_plot_area('Fertility Rate of Age Group 15-19', tfr_15_19, [colors_15_19[0], colors_15_19[-1]])
make_plot_area('Fertility Rate of Age Group 20-24', tfr_20_24, [colors_20_24[0], colors_20_24[-1]])
make_plot_area('Fertility Rate of Age Group 25-29', tfr_25_29, [colors_25_29[0], colors_25_29[-1]])
make_plot_area('Fertility Rate of Age Group 30-34', tfr_30_34, [colors_30_34[0], colors_30_34[-1]])
make_plot_area('Fertility Rate of Age Group 35-39', tfr_35_39, [colors_35_39[0], colors_35_39[-1]])
make_plot_area('Fertility Rate of Age Group 40-44', tfr_40_44, [colors_15_19[0], colors_15_19[-1]])
make_plot_area('Fertility Rate of Age Group 45-49', tfr_45_49, [colors_20_24[0], colors_20_24[-1]])
```

Here we can see that the women of urban area have lower fertility rate for all age groups since , they tend to be more literate than the women of rural area , and thus the pattern in the data .
